```python
import ollama
from IPython.display import display, Markdown, clear_output

client = ollama.Client(host='http://host.docker.internal:11434')
```

```python
def chat(message):
    stream = client.chat(
        model='gemma4:31b',
        messages=[
            {'role': 'user', 'content': message}
        ],
        stream=True
    )

    full_response = ''
    for chunk in stream:
        full_response += chunk['message']['content']
        clear_output(wait=True)
        display(Markdown(full_response))
```

```python
chat("Tell me about the japanese word tengoku")
```

```python
import base64
from pathlib import Path
from IPython.display import display, Markdown, clear_output, Image as IPImage

image_path = 'something.png'  # update this

# Display the image inline before sending
#display(IPImage(filename=image_path))

with open(image_path, 'rb') as f:
    image_b64 = base64.b64encode(f.read()).decode()

stream = client.chat(
    model='gemma4:31b',
    messages=[{
        'role': 'user',
        'content': 'Rank this art on a numberic scale for me. 1-100. I will use the number to show it in a queue of images and I want the best images first in the list. RESPOND WITH A NUMBER ONLY!',
        'images': [image_b64]
    }],
    stream=True
)

full_response = ''
for chunk in stream:
    full_response += chunk['message']['content']
    clear_output(wait=True)
    #display(IPImage(filename=image_path))
    display(Markdown(full_response))
```
